Optimisation With Optuna

```python
def objective(trial):
    params = {
        'objective'        : 'binary:logistic',
        'n_estimators'     : 500,
        'random_state'     : 42,
        'n_jobs'           : -1,
        'scale_pos_weight' : scale,
        'learning_rate'    : trial.suggest_float('learning_rate', 0.01, 0.2, log=False),
        'max_depth'        : trial.suggest_int('max_depth', 3, 12),
        'min_child_weight' : trial.suggest_float('min_child_weight', 1, 10),
        'subsample'        : trial.suggest_float('subsample', 0.6, 1.0),
        'colsample_bytree' : trial.suggest_float('colsample_bytree', 0.6, 1.0),
        'reg_alpha'        : trial.suggest_float('reg_alpha', 1e-8, 10.0, log=False),
        'reg_lambda'       : trial.suggest_float('reg_lambda', 1e-8, 10.0, log=False),
    }
    
    model = XGBClassifier(**params)
    model.fit(X_train_original, y_train_original, eval_set=[(X_val, y_val)])
    
    y_pred = model.predict(X_val)
    score  = f1_score(y_val,y_pred)

    return score


study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=50, show_progress_bar=True)

print("Number of completed trials : ", len(study.trials))
print("Best one : ")
trial = study.best_trial

print("  Value (F1-Score)    : ", trial.value)
print("  Best settings : ")
for key, value in trial.params.items():
    print(f"    {key} : {value}")

best_params = trial.params
```

```python
def objective_smote(trial):
    params = {
        'objective': 'binary:logistic',       
        'n_estimators': 100,                   
        'random_state': 42,
        'n_jobs': -1,
        
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=False),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'min_child_weight': trial.suggest_float('min_child_weight', 1e-3, 10.0, log=False),
        'gamma': trial.suggest_float('gamma', 0.0, 10.0, log=False),  
        'subsample': trial.suggest_float('subsample', 0.6, 1.0),     
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.6, 1.0),
        'reg_alpha': trial.suggest_float('reg_alpha', 1e-8, 10.0, log=False),   # L1 regulzarization
        'reg_lambda': trial.suggest_float('reg_lambda', 1e-8, 10.0, log=False), # L2 regularization
        'scale_pos_weight': trial.suggest_float('scale_pos_weight', 0.5, 5.0) 
    }
    
    model = XGBClassifier(**params)
    
    model.fit(X_train_resampled, y_train_resampled, eval_set=[(X_val, y_val)]) 
    
    y_pred = model.predict(X_val)
    score = f1_score(y_val, y_pred)

    return score

print("n--- Launching Optuna optimization for the SMOTE model ---\\n")
study_smote = optuna.create_study(direction='maximize', pruner=optuna.pruners.MedianPruner())
study_smote.optimize(objective_smote, n_trials=50, show_progress_bar=True) 

print("\\n--- Best hyperparameters found for SMOTE ---")
best_params_smote = study_smote.best_trial.params
for key, value in best_params_smote.items():
    print(f"    {key}: {value}")

# Entraînement du modèle final optimisé AVEC SMOTE
print("\\n--- Training the final optimized model with SMOTE ---")
XGboost_best = XGBClassifier(
    n_estimators=1000,
    random_state=42,
    n_jobs=-1,
    **best_params_smote
)

start_time = time()
XGboost_best.fit(X_train_resampled, y_train_resampled)
training_time_smote = time() - start_time
minutes = int(training_time_base // 60)  
seconds = training_time_base % 60 

print(f"Trained in "
    + Fore.LIGHTCYAN_EX + Style.BRIGHT + f"{int(minutes)}" + Style.RESET_ALL + " min "
    + Fore.LIGHTCYAN_EX + Style.BRIGHT + f"{int(seconds)}" + Style.RESET_ALL + " seconds\n"
)
# Évaluation sur le jeu de validation
y_pred_smote_best = XGboost_best.predict(X_val)
y_pred_proba_smote_best = XGboost_best.predict_proba(X_val)[:, 1]

print("--- Performance on the validation set (Optimized SMOTE model) ---")
print(f"F1-Score:          {f1_score(y_val, y_pred_smote_best) * 100:.4f} %")
print(f"Precision:         {precision_score(y_val, y_pred_smote_best) * 100:.4f} %")
print(f"Recall:            {recall_score(y_val, y_pred_smote_best) * 100:.4f} %")
print(f"ROC-AUC:           {roc_auc_score(y_val, y_pred_proba_smote_best) * 100:.4f} %")
print(f"PR-AUC:            {average_precision_score(y_val, y_pred_proba_smote_best) * 100:.4f} %\\n")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import f1_score, roc_auc_score, log_loss
from xgboost import XGBClassifier

model = XGBClassifier(
    n_estimators=5,      
    learning_rate=0.1,
    max_depth=5,
    random_state=42,
    use_label_encoder=False
)

f1_scores = []
auc_scores = []
logloss_scores = []

for i, y_pred in enumerate(model.staged_predict(X_val) if hasattr(model, "staged_predict") else range(1, 201)):
    model.n_estimators = i+1
    model.fit(X_train_resampled, y_train_resampled)
    
    y_pred_val = model.predict(X_val)
    y_pred_proba_val = model.predict_proba(X_val)[:,1]
    
    f1_scores.append(f1_score(y_val, y_pred_val))
    auc_scores.append(roc_auc_score(y_val, y_pred_proba_val))
    logloss_scores.append(log_loss(y_val, y_pred_proba_val))


plt.figure(figsize=(12,6))
plt.plot(range(1, len(f1_scores)+1), f1_scores, label='F1-Score', color='blue')
plt.plot(range(1, len(auc_scores)+1), auc_scores, label='ROC-AUC', color='green')
plt.plot(range(1, len(logloss_scores)+1), logloss_scores, label='Log Loss', color='red')
plt.xlabel("Number of trees (n_estimators)")
plt.ylabel("Score / Loss")
plt.title("Performance evolution according to the number of trees")
plt.legend()
plt.grid(True)
plt.show()
```

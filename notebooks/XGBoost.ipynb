```python
def print_header(title):
    color = Fore.LIGHTMAGENTA_EX + Style.BRIGHT
    color_text = Fore.LIGHTCYAN_EX + Style.BRIGHT
    line  = "~" * (len(title) + 10)
    print(color + line)
    print(color_text + f"~~~  {title}  ~~~")
    print(color + line + Style.RESET_ALL + "\n")

X_train_original = X_train_full
y_train_original = y_train_full
X_train_smote    = X_train_resampled
y_train_smote    = y_train_resampled

def train_and_evaluate(X_train, y_train, X_val, y_val, label):
    print_header(f"XGBoost Model ({label})")


    model = XGBClassifier(
        n_estimators      = 1000,
        seed              = 123,
        learning_rate     = 0.1,
        max_depth         = 6,
        num_leaves        = 31,
        scale_pos_weight  = scale,
        min_child_samples = 20,
        min_child_weight  = 0.001,
        subsample         = 0.8,
        colsample_bytree  = 0.8,
        reg_alpha         = 0.0,
        reg_lambda        = 0.0,
        n_jobs            = -1,
        random_state      = 42,
    )

    start_time = time()
    model.fit(X_train, y_train)
    training_time = time() - start_time

    minutes = int(training_time // 60)
    seconds = training_time % 60
    print(f"Trained in {Fore.LIGHTCYAN_EX}{minutes}{Style.RESET_ALL} min "
          f"{Fore.LIGHTCYAN_EX}{int(seconds)}{Style.RESET_ALL} sec\n")

    y_pred = model.predict(X_val)
    
    print_header("Confusion Matrix")
    cm = confusion_matrix(y_val, y_pred_base)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Legit', 'Fraud'])
    disp.plot(cmap='Blues')
    plt.title('Confusion Matrix (Val Set)')
    plt.show()

    print_header("classification_report")
    print(classification_report(y_val, y_pred))
    print_header("Accuracy")
    print(f"Accuracy : {accuracy_score(y_val, y_pred):.2f}\n")

    return model, training_time
```

WITHOUT SMOTE

```python
model_original, time_original = train_and_evaluate(X_train_original, y_train_original, X_val, y_val, "Whithout SMOTE")
```

WITH SMOTE

```python
model_smote, time_smote = train_and_evaluate(X_train_smote, y_train_smote, X_val, y_val, "With SMOTE")
```

COMPARISON

```python
print_header("Analyse between both models")
print(f"Without SMOTE : {time_original:.2f} s")
print(f"With SMOTE    : {time_smote:.2f} s\n")
```

RANDOM FOREST METHOD

```python
import pandas  as pd 
import numpy   as np 
import matplotlib.pyplot   as plt 
import seaborn  as sns 
from sklearn.ensemble  import RandomForestClassifier
from sklearn.metrics   import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score, average_precision_score, confusion_matrix, classification_report, ConfusionMatrixDisplay, roc_curve, auc
from colorama   import Fore, Style
from time import time
```

```python
X_train_original = X_train_full
y_train_original = y_train_full

X_train_smote    = X_train_resampled
y_train_smote    = y_train_resampled


print(f"`\n{'~'*10} Random Forest Method {'~'*10}\n")
print(f"Original train : {X_train_original.shape}")
print(f"SMOTE train    : {X_train_smote.shape}")
print(f"Validation     : {X_val.shape}\n")



def train_and_evaluate_rf(X_train, y_train, X_val, y_val, label):
    print(f"\n{'~'*10}Random Forest Model ({label}) :{'~'*10}\n")

    # Initialize RandomForest Model
    rf_model = RandomForestClassifier(
        n_estimators=5,
        max_depth=None,
        min_samples_split=2,
        min_samples_leaf=1,
        max_features='sqrt',
        class_weight='balanced',
        n_jobs=-1,
        random_state=42
    )

    # Training
    start_time = time()
    rf_model.fit(X_train, y_train)
    training_time = time() - start_time

    minutes = int(training_time // 60)
    seconds = training_time % 60
    print(f"Trained in {Fore.LIGHTCYAN_EX}{minutes}{Style.RESET_ALL} min "
          f"{Fore.LIGHTCYAN_EX}{int(seconds)}{Style.RESET_ALL} sec\n")

    # Prédictions
    y_pred = rf_model.predict(X_val)
    y_pred_proba = rf_model.predict_proba(X_val)[:,1]


    print(f"\n{'~'*10} Performance Metrics :{'~'*10}\n")
    print(f"F1-Score:          {f1_score(y_val, y_pred)*100:.2f} %")
    print(f"Precision:         {precision_score(y_val, y_pred)*100:.2f} %")
    print(f"Recall:            {recall_score(y_val, y_pred)*100:.2f} %")
    print(f"ROC-AUC:           {roc_auc_score(y_val, y_pred_proba)*100:.2f} %")
    print(f"PR-AUC:            {average_precision_score(y_val, y_pred_proba)*100:.2f} %\n")


    print(f"\n{'~'*10} Confusion Matrix : {'~'*10}\n")
    cm = confusion_matrix(y_val, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Legit', 'Fraud'])
    disp.plot(cmap='Blues')
    plt.title(f'Confusion Matrix ({label})')
    plt.show()


    print(f"\n{'~'*10} Classification Report : {'~'*10}\n")
    print(classification_report(y_val, y_pred))

    print(f"\n{'~'*10} Accuracy : {accuracy_score(y_val, y_pred):.2f} {'~'*10}\n")

    print(f"\n{'~'*10} Top Features : {'~'*10}\n")
    feature_importances = pd.Series(rf_model.feature_importances_, index=X_train.columns)
    feature_importances.nlargest(10).plot(kind='barh', figsize=(10,6))
    plt.title(f'Top 10 Features - Random Forest ({label})')
    plt.show()

    return rf_model, training_time
```

Without SMOTE without optimization

```python
model_original, time_original = train_and_evaluate_rf(X_train_original, y_train_original, X_val, y_val, "Without SMOTE")
```

With SMOTE without optimization

```python
model_smote, time_smote       = train_and_evaluate_rf(X_train_smote, y_train_smote, X_val, y_val, "With SMOTE")
```

```python
print(f"\n{'~'*10} Comparaison between With and Without SMOTE {'~'*10}\n")
print(f"Without SMOTE : {time_original:.2f} s")
print(f"With SMOTE    : {time_smote:.2f} s\n")
```

Optimization for the random forest method

```python
print(f"\n{'~'*10}Optuna Hyperparameter Optimization - Random Forest {'~'*10}\n ")

import optuna
from sklearn.model_selection import cross_val_score

def objective(trial):
    #hyperparameters
    params = {
        "n_estimators": trial.suggest_int("n_estimators", 2, 5),
        "max_depth": trial.suggest_int("max_depth", 2, 7),
        "min_samples_split": trial.suggest_int("min_samples_split", 2, 20),
        "min_samples_leaf": trial.suggest_int("min_samples_leaf", 1, 20),
        "max_features": trial.suggest_categorical("max_features", ["sqrt", "log2", None]),
        "bootstrap": trial.suggest_categorical("bootstrap", [True, False])
    }

    # Model
    model = RandomForestClassifier(
        **params,
        class_weight='balanced',
        n_jobs=-1,
        random_state=42
    )

    scores = cross_val_score(model, X_train_original, y_train_original, cv=5, scoring='accuracy')

    return scores.mean()


study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=15, show_progress_bar=True)

print(f"number of attempts : {len(study.trials)}")
print(f"Best score : {study.best_value:.4f}")
print(f"Best hyperparameters :")
for key, value in study.best_params.items():
    print(f"   {key}: {value}")
```

Optimization training and evaluation

```python
def train_and_evaluate_rf(X_train, y_train, X_val, y_val, label, best_params):
    print(f"\n{'~'*10} Random Forest optimization ({label}) {'~'*10}\n")
    
    model = RandomForestClassifier(
        **best_params,
        class_weight='balanced',
        n_jobs=-1,
        random_state=42
    )

    start_time = time()
    model.fit(X_train, y_train)
    train_time = time() - start_time

    y_pred = model.predict(X_val)
    y_proba = model.predict_proba(X_val)[:,1]

    print(f"\n{'~'*10}Training time: {train_time:.2f} sec{'~'*10}\n")

    print(f"\n{'~'*10} Performance Metrics :{'~'*10}\n")
    print(f"Accuracy:      {accuracy_score(y_val, y_pred)*100:.2f} %")
    print(f"F1-score:      {f1_score(y_val, y_pred)*100:.2f} %")
    print(f"Precision:     {precision_score(y_val, y_pred)*100:.2f} %")
    print(f"Recall:        {recall_score(y_val, y_pred)*100:.2f} %")
    print(f"ROC-AUC:       {roc_auc_score(y_val, y_proba)*100:.2f} %")
    print(f"PR-AUC:        {average_precision_score(y_val, y_proba)*100:.2f} %\n")

    # Confusion matrix
    print(f"\n{'~'*10} Confusion Matrix : {'~'*10}\n")
    cm = confusion_matrix(y_val, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Legit', 'Fraud'])
    disp.plot(cmap='Blues')
    plt.title(f'Confusion Matrix ({label})')
    plt.show()


    print(f"\n{'~'*10} Classification Report : {'~'*10}\n")
    print(classification_report(y_val, y_pred))

    print(f"\n{'~'*10} Accuracy : {accuracy_score(y_val, y_pred):.2f} {'~'*10}\n")

    print(f"\n{'~'*10} Top Features : {'~'*10}\n")
    feature_importances = pd.Series(model.feature_importances_, index=X_train.columns)
    feature_importances.nlargest(10).plot(kind='barh', figsize=(10,6))
    plt.title(f'Top 10 Features - Random Forest ({label})')
    plt.show()

    return model, train_time


best_params = study.best_params  #retrieve best hyperparameters from the optimization
```

Without SMOTE with optimization

```python
model_no_smote, t_no_smote = train_and_evaluate_rf(X_train_original, y_train_original, X_val, y_val, "Without SMOTE", best_params)
```

With SMOTE with optimization

```python
model_smote, t_smote = train_and_evaluate_rf(X_train_smote, y_train_smote, X_val, y_val, "With SMOTE", best_params)
```
